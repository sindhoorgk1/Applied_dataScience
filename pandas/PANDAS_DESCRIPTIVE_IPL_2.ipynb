# Groupby and aggregating(without series and by creating dataframe)

```python
sold_price_by_age = ipldf.groupby('AGE')['SOLD PRICE'].mean().reset_index()
```

```python
sold_price_by_age
```

# Grouping and aggregating on Multiple columns

```python
sold_price_by_age_role = ipldf.groupby(['AGE','PLAYING ROLE'])['SOLD PRICE'].mean().reset_index()
sold_price_by_age_role
```

# Joining Dataframe:inner,outer,left,right

```python
soldprice_comparison=sold_price_by_age_role.merge(sold_price_by_age,on = 'AGE',how = 'outer')
```

```python
soldprice_comparison
```

# Renaming the Columns

```python
soldprice_comparison.rename(columns = {'SOLD PRICE_x':'SOLD_PRICE_AGE_ROLE','SOLD PRICE_y':'SOLD_PRICE_AGE'},inplace = True)
```

```python
soldprice_comparison.head()
```

# Applying operations on multiple columns(apply() on each row with calculations)

```python
soldprice_comparison['CHANGE']=soldprice_comparison.apply(lambda rec:(rec.SOLD_PRICE_AGE_ROLE-rec.SOLD_PRICE_AGE)/rec.SOLD_PRICE_AGE,axis = 1)
```

```python
soldprice_comparison
```
